```python
def my_plotter(x_list, y_list, error_list, color_list, label_list, save_name):
    index = 0
    plt.hlines(1.0, x_list[0][0], x_list[0][-1], linestyles='dashed')
    for i in range(len(y_list)):
        x = x_list[i]
        y = y_list[i]
        error = error_list[i]
        label = label_list[i]
#         error = np.std(y)
        avgx = np.sum(x)/x.shape[0]
        normalizedx = avgx / avgx
        plt.plot(x, y, 'k', color=color_list[index], label=label)
        plt.fill_between(x, y-error, y+error,
            alpha=0.2, edgecolor=color_list[index], facecolor=color_list[index],
            linewidth=1, antialiased=True)
#         index = len(color_list) - np.mod(len(color_list), index + 1)
        index += 1
    plt.axis('on')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Compression Ratio vs Baseline')
    plt.legend()
#     plt.gca().set_position([0, 0, 1, 1])
    plt.savefig(save_name, dpi=500, quality=95)
    plt.show()
```

```python
x_list = [cluster_array, cluster_array, cluster_array, cluster_array]
y_list = [normalized4_2[:,0], normalized4_2[:,2], normalized4_2r[:,0], normalized4_2r[:,2]]
error_list = [(np.std(results4_2, axis=0)/ results4_2b)[0,:,0],
              (np.std(results4_2, axis=0)/ results4_2b)[0,:,2],
              (np.std(results4_2r, axis=0)/ results4_2b)[0,:,0],
              (np.std(results4_2r, axis=0)/ results4_2b)[0,:,2]]
color_list = ['#003f5c', '#7a5195','#ef5675', '#ffa600']
label_list = ['Kmeans (Train Set)', 'Kmeans (Test Set)', 'Random (Train Set)', 'Random (Test Set)']
my_plotter(x_list, y_list, error_list, color_list, label_list, "results4.jpeg")
```
